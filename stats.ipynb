### **Intro**

In multiple linear regression, we attempt to explain the variation in a dependent response variable based on the variation in a set of independent predictor variables. How can narrow in on what predictors are the most effective, while also maximizing simplicity according to *Occam's razor*? **How can we balance fitting and generalization?**

Let's say we want to predict $y$ = `number_of_yells` based on $X_1$ = `football_games`, $X_2$ = `basketball_games`, $X_3$ = `volleyball_games`, and $X_4$ = `baseball_games`.

\begin{align}
 y = \beta_0 + \beta_1x_1 + \beta_2x_2 + \beta_3x_3 + \beta_4x_4
\end{align}

However, our intuition tells us that most of the yells you learn are probably from the football games, so a simpler model may be more intuitive.

\begin{align}
 y = \beta_0 + \beta_1x_1
\end{align}

When evaluating a model, the simplest measure to look at is correlation $r$. When we get the correlation for the 2 models above, we'll see that the correlation is higher for the first model, even though 3 of the predictors might not impact the response. **In general, more predictors will always increase the correlation coefficient $r$ due to the simplicity of the measure.**
*   If the predictor has no impact ($\beta_{new} = 0$), then the correlation will be the same.
*   If the predictor has any impact ($\beta_{new} \neq 0$), then the correlation increases.

**Primary Question: Is the extra variation in response explained by additional predictors *significant enough* to warrant a more complex model?**

### **Adjusted $R^2$**



R-squared and Adjusted R-squared both evaluate the goodness-of-fit of a regression model. They provide insights into how well the model explains the variability in the data.

Unlike the standard R-squared, which simply tells you the proportion of variance explained by the model, **adjusted R-squared accounts for the number of predictors in the model**. More specifically, adjusted R-squared *penalizes a higher number of predictors*, so adding a new predictor will only increase the metric if there is a significant improvement in model performance.

\begin{align}
  R^2_{adj} = 1 - \frac{(1 - R^2)(N - 1)}{N - p - 1}
\end{align}

where $R^2$ is the standard correlation of the model, $N$ is the sample size, and $p$ is the number of predictors.

**Therefore, adjusted R-squared balances model performance with generalizability**, making it a valuable tool for model selection.

```python
y = data["number_of_yells"]
X = data[["football_games", "basketball_games", "volleyball_games", "baseball_games"]]

ols = sm.OLS(y, X)
ols_fit = ols.fit()

full_rsquared = ols_fit.rsquared
full_rsquared_adj = ols_fit.rsquared_adj

print_str = f"""Full Model
  R-squared = {full_rsquared}
  Adjusted R-squared = {full_rsquared_adj}
"""
print(print_str)
```

### **Quick Info about Hypothesis Tests**

A hypothesis test shows whether there is significant evidence of a given null hypothesis. Ultimately, the decision comes down to the statistician, who interprets the results.

We have a null hypothesis $H_0$, which is often what we are trying to disprove (not always). The alternative hypothesis $H_a$ is accepted if we reject $H_0$. An example is below:

> $H_0: $ Number of football games does not affect whether you're a 2-percenter

> $H_a: $ Number of football games has some effect on whether you're a 2-percenter

In the hypothesis test, we assume $H_0$ is true and find the probability that the sample would look like what it does. In other words, how likely is it that we will get this sample distribution if $H_0$ is true?

If the probability is high, then we fail to reject $H_0$ since it's entirely possible to get the sample values. If the probability is low, then we reject $H_0$ and accept $H_a$.

### **Model Selection Tests**

A test of significance can help us determine whether we should go with a complex model or a simpler one.

#### **t-test**

Using a t-test, we can test whether a given predictor $x_i$ is insignificant ($\beta_i = 0$). The most common hypotheses are below, but the test can also be one-tailed ($\beta_i > 0$ or $\beta_i < 0$).

\begin{align}
  H_0: \beta_i = 0 \\
  H_a: \beta_i \neq 0
\end{align}

The t-test assumes all other variables are significant, so it can only test a single predictor at a time and holds all others constant. But in our situation with `number_of_yells`, we're interested in removing multiple predictors. How do we test this?

#### **F-test**

Using an F-test, we can test whether a given set of predictor variables are significant in explaining the variance of the dependent variable.

Let's say we have a full model $y = \beta_0 + \beta_1x_1 + ... + \beta_px_p$, where $p$ is the number of predictors. We will want to test the hypothesis:




> $H_0: \beta_{i_1} = 0, \beta_{i_2} = 0, ..., \beta_{i_k} = 0$ \\

> $H_a: $ At least one $\beta_{i_j} \neq 0$


Therefore, we can test a reduced model versus the full model and see if the additional parameters are significant in determining the response.

We used the Ordinary Least Squares (OLS) functionality from `statsmodels.api` to see a specific metric, but we can also see an entire summary of the metrics.

```python
y = data["number_of_yells"]
X = data[["football_games", "basketball_games", "volleyball_games", "baseball_games"]]

ols = sm.OLS(y, X)
ols_fit = ols.fit()
ols_fit.summary()
```

From the full model, we can reduce the model is various ways with different null hypotheses, as shown below. To do the F-test in Python, we just have to set up a matrix describing our $H_0$.

Based on the results below, the p-value is very high for the F-test where we exclude all predictors except football games, which means we fail to reject to null. **Therefore, we can reduce the model to only use the predictor `football_games`.**

```python
# Null Hypothesis: b4 = 0
A = np.array(([0, 0, 0, 1]))

print(ols_fit.f_test(A))
print()
```

## Correlation vs Cramer's V


1. Context of Use:

- Correlation: Used for continuous variables.
- Cramer's V: Used for categorical variables.

2. Purpose:

- Correlation: Measures the strength and direction of a linear relationship between two continuous variables.
- Cramer's V: Measures the strength of association between two categorical variables.

3. Value Range:

- Correlation: Ranges from -1 to +1, where -1 indicates a perfect negative linear relationship, +1 indicates a perfect positive linear relationship, and 0 means no linear correlation.
- Cramer's V: Ranges from 0 to 1, where 0 indicates no association and 1 indicates a perfect association.

4. Assumptions:

- Correlation (specifically Pearson's correlation): Assumes a linear relationship and normally distributed data.
- Cramer's V: Does not assume a linear relationship or normal distribution, suitable for nominal and ordinal data.

## Cramer's V


Cramer's V is a statistical measure used to assess the strength of association between two categorical variables. It ranges from 0 to 1, where 0 indicates no association and 1 indicates a perfect association. The value is derived from a Chi-square statistic and is suitable for nominal and ordinal data. Cramer's V is particularly useful because it provides a standardized measure of association that is not affected by the size of the table or the number of categories in the variables.

## Errors

Imagine you run a lemonade stand and you're trying to predict how many cups of lemonade you'll sell each day.

1. Predictions vs Actual Sales:

- You predict that on Monday, you'll sell 30 cups, but you actually sell 35 cups.
- On Tuesday, your prediction is 40 cups, but you sell 38 cups.
- On Wednesday, you predict 50 cups, and you sell exactly 50 cups.

2. Calculate Errors:

- The error for Monday is 30 (predicted) - 35 (actual) = -5 cups.
- For Tuesday, it's 40 - 38 = 2 cups.
- And for Wednesday, it's 50 - 50 = 0 cups.

3. Square Each Error:

- Squaring -5 gives 25 (since -5 × -5 = 25).
- Squaring 2 gives 4.
- Squaring 0 gives 0.

4. Average of Squared Errors:

- Add up these squared errors: 25 + 4 + 0 = 29.
- Since there are 3 days, the average is 29 / 3 ≈ 9.67.

So, the Mean Squared Error for your lemonade sales predictions is about 9.67. This number represents how off your predictions were on average, with a smaller number indicating better accuracy. In this case, a MSE of 9.67 means your predictions were relatively close to the actual sales.

The Sum Squared Errors would be 29 giving you a magnitude instead

**When would you use either?**
